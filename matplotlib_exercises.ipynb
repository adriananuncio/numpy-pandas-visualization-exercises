##### 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
a. y = âˆšx

b. y = x^3

c. y = 2^x

d. y = 1/(x+1)

You can use functions from the math module to help implement some of the equations above.

```python
# a. y = âˆšx

x = np.linspace(0, 50)
y = np.sqrt(x)

plt.plot(x,y)
plt.title('Square root of X')
plt.grid(color='grey', linestyle='-', linewidth='0.05')
plt.text(0,0, '(0,0)', 
          fontsize = 8 , 
          color= 'r')
```

```python
# b. y = x^3

x = np.linspace(0,10)
y = x**3

plt.plot(x,y)
plt.title('X Cubed')
plt.text(0,0, '(0,0)', 
          fontsize = 8 , 
          color= 'r')
```

```python
# c. y = 2^x

x = np.linspace(0,10)
y = 2**x

plt.plot(x,y)
plt.title('2 Squared')
plt.text(0,0, '(0,0)', 
          fontsize = 8 , 
          color= 'r')
```

```python
# d. y = 1/(x+1)

x = np.linspace(-1,10)
x2= np.linspace(-10,-1)
y = np.reciprocal(x+1)
y2= np.reciprocal(x2+1)

plt.plot(x,y, 'b')
plt.plot(x2,y2, 'b')
plt.title('Reciprocal Function')
plt.text(0,0, '(0,0)', 
          fontsize = 8 , 
          color= 'r')
```
