```python
# imports

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
# 1. What does the distribution of petal lengths look like?
# Distibution of petal lengths is right skewed

petal_len_dist = sns.displot(data=iris, x='petal_length')

petal_len_dist
```

```python
# 2. Is there a correlation between petal length and petal width?
# There is a strong positive correlation between petal length and petal width

sns.relplot(data=iris, x='petal_length', y='petal_width')
plt.show()
```

```python
# 3. Would it be reasonable to predict species based on 
# sepal width and sepal length? For this, you'll visualize 
# two numeric columns through the lense of a categorical column.
plt.suptitle('Sepal length and width by species')

plt.subplot(121)
sns.boxplot(data=iris, x='species', y='sepal_length')
plt.title('Sepal length by species')

plt.subplot(122)
sns.boxplot(data=iris, x='species', y='sepal_width')
plt.title('Sepal width by species')

plt.show()
```

```python
# It would not be a good idea to predict species based 
# on sepal length and sepal width because there is 
# not a significant difference in their values.
```

```python
# 4. Which features would be best used to predict species?
sns.heatmap(iris.corr(),
    annot=True,
    cmap='coolwarm',
    vmax=1,
    vmin=-1,
    center=0,
    linewidth=1,
    linecolor='k',
    fmt='.2f')

plt.show()
```

```python
iris.corr()
```

```python
sns.pairplot(iris, corner = True)
plt.show()
```

```python
plt.suptitle('Petal length and width by species')

plt.subplot(121)
sns.boxplot(data=iris, x='species', y='petal_length')
plt.title('Petal length by species')

plt.subplot(122)
sns.boxplot(data=iris, x='species', y='petal_width')
plt.title('Petal width by species')

plt.show()
```

```python
# petal length and width would be better features to 
# make species predictions with because there is a strong 
# positive corrilation indicated by the 0.96 correlation value 
# and their box plots show a more significant difference in values
```

```python
# 3. Load the swiss dataset and read it's documentation. 
# The swiss dataset is available from pydatset rather than 
# seaborn. Create visualizations to answer the following 
# questions:

from pydataset import data

df = data('swiss')

df.head()
```

```python
# 3a. Create an attribute named is_catholic that holds a 
# boolean value of whether or not the province is Catholic. 
# (Choose a cutoff point for what constitutes catholic)

df['is_catholic'] = np.where(df.Catholic >= 70.00, 1, 0)
```

```python
df.head()
```

```python
# 3b. Does whether or not a province is Catholic influence 
# fertility?
sns.boxplot(data=df, x='is_catholic', y='Fertility')
plt.show()

df.corr()
# 0 = not catholic, 1 = catholic
```

##### The correlation value, 0.605, indicats a slight linear correlation between being catholic and fertility. The box plots show significant overlap however, suggesting no correlation.

```python
# 3c. What measure correlates most strongly with fertility?

sns.heatmap(df.corr(),
           annot=True,
           cmap='coolwarm',
           vmax=1,
           vmin=-1,
           center=0,
           linewidth=1,
           linecolor='k',
           fmt='.2f')
plt.show()
```

##### Education most strongly correlates to fertility. It is a negative correlation, slightly linear.
